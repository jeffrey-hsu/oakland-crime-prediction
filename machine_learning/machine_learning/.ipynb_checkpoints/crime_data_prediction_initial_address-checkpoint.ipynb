```python
# Default API key: replace with your own api key
google_api_key = "AIzaSyAvErouO-rPBPtjFmzuZJ7c2E5hPYQDQi0"

# Function to call Google's geocode api based on an address
# and returns a dictionary in json format
def get_geocode_json(address, api_key):
    geo_s ='https://maps.googleapis.com/maps/api/geocode/json'
    param = {'address': address, 'key': google_api_key }    
    response = requests.get(geo_s, params=param)
    result_dict = response.json()
    return result_dict    
```

```python
# this code essentially proves that all the examples in addresses column are not strings
address_len = len(crime_data['Address'])
np.sum([isinstance(crime_data.iloc[i]['Address'],str) for i in range(address_len)])
```

```python
# this will hopefully give us all the valid addresses
strt_addresses = []

for i in range(address_len):
    if isinstance(crime_data.iloc[i]['Address'],str):
        strt_addresses.append(crime_data.at[crime_data.index[i], 'Address'] + ", " + crime_data.at[crime_data.index[i], 'City'] + ", " + crime_data.at[crime_data.index[i], 'State'])
```

```python
# we are only recording the unique addresses
address = list(set(strt_addresses))
address[0:30]
```

```python
lt = len(address)
print(lt)
```

```python
Lat = [''] * lt
Long = [''] * lt
Zip = [''] * lt
```

```python
for i in range(len(address)):
    print("Iteration i = " + str(i))
    
    result_dict = get_geocode_json(address[i], google_api_key)
    
    print("len(result_dict['results']) = %d"%(len(result_dict['results'])))
    try:
        Lat[i] = result_dict['results'][0]['geometry']['location']['lat']
    except:
        print("i = " + str(i) + ": Cannot find Latitude")

    
    try:
        Long[i] = result_dict['results'][0]['geometry']['location']['lng']
    except:
        print("i = " + str(i) + ": Cannot find Longitude")
        
    try:    
        for j in range(len(result_dict['results'][0]['address_components'])):
            if result_dict['results'][0]['address_components'][j]['types'][0] == 'postal_code':
                Zip[i] = result_dict['results'][0]['address_components'][j]['long_name']
    except:
        print("i = " + str(i) + ": Cannot find Address Components")
          
```

```python
# we will try to impute values based on zipcode of places with closest Lat and Long
address_nomiss = address_dataset.loc[((address_dataset['Long'] != '') & (address_dataset['Lat'] != '')) & (address_dataset['Zipcode'] != '')]
```

```python
for i in range(len(address_zipmiss['Zipcode'])):
    print("iteration: " + str(i))
    diff = 5.0
    index = 20000
    for j in range(len(address_nomiss['Lat'])):
        d = np.absolute(address_zipmiss.iloc[i]['Lat'] - address_nomiss.iloc[j]['Lat']) + np.absolute(address_zipmiss.iloc[i]['Long'] - address_nomiss.iloc[j]['Long']) 
        if d < diff:
            diff = d
            index = j
    if index != 20000:
        address_zipmiss.iloc[i]["Zipcode"] = address_nomiss.iloc[index]['Zipcode']       
```
