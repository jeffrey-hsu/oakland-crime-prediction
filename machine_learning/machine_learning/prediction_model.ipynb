```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# fitting training data for No. of principle components equal to 2
pca = PCA(n_components = 2)
X_train = pca.fit_transform(X)
    
# fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 0)
y_kmeans = kmeans.fit(X_train)    
```

```python
type(y_kmeans)
```

```python
# finding the unique set of labels 
label_set = set(y_kmeans.labels_)
    
# finding distance of each 2d point for the centroid of its assigned cluster
dist_cent = []
for i in range(0,len(y_kmeans.labels_)):
    dist_cent.append(np.linalg.norm(X_train[i,]-y_kmeans.cluster_centers_[y_kmeans.labels_[i]]))
```

```python
# finding the distance of the most distant point from the centroid of its assigned cluster
max_dist_cent = []
max_dist_cent_x = []
max_dist_cent_y = []
for n in label_set:
    n_val = []
    for i in list(np.where(y_kmeans.labels_==n)[0]):
        n_val.append(dist_cent[i])
    max_dist_cent.append(max(n_val))
```

```python
# plotting the circles with cluster centroids as centers and radius equal to distance of
# centroid from the most distant point assigned to that cluster
circles = []
for i in range(0,4):
    circles.append(plt.Circle((y_kmeans.cluster_centers_[i,0], y_kmeans.cluster_centers_[i,1])
                                  ,max_dist_cent[i], color='g', clip_on=False,
                                  linewidth=3.0,alpha =0.3))
```

```python
fig=plt.figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
plt.plot(X_train[:,0], X_train[:,1],'bs',alpha=0.35, markersize=7.5)
plt.axis([min(X_train[:,0])-2,max(X_train[:,0])+2,min(X_train[:,1])-2,max(X_train[:,1])+2])
plt.scatter(y_kmeans.cluster_centers_[:,0], y_kmeans.cluster_centers_[:,1],color = 'black')
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("The "+ str(4) + " Identified Clusters")
ax=fig.add_subplot(1,1,1)
for i in range(0,4):
    ax.annotate("C " + str(i+1),(y_kmeans.cluster_centers_[i,0],y_kmeans.cluster_centers_[i,1]+0.1)) 
plt.show()
```

```python
# fitting training data for No. of principle components equal to 2
n_clus = 8
pca = PCA(n_components = 2)
X1_train = pca.fit_transform(X1)
    
# fitting K-Means to the dataset
kmeans = KMeans(n_clusters = n_clus, init = 'k-means++', random_state = 0)
y1_kmeans = kmeans.fit(X1_train)

# finding the unique set of labels 
label_set = set(y1_kmeans.labels_)
    
# finding distance of each 2d point for the centroid of its assigned cluster
dist1_cent = []
for i in range(0,len(y1_kmeans.labels_)):
    dist1_cent.append(np.linalg.norm(X1_train[i,]-y1_kmeans.cluster_centers_[y1_kmeans.labels_[i]]))
    
# finding the distance of the most distant point from the centroid of its assigned cluster
max_dist_cent = []
for n in label_set:
    n_val = []
    for i in list(np.where(y1_kmeans.labels_==n)[0]):
        n_val.append(dist1_cent[i])
    max_dist_cent.append(max(n_val))
    
mark = ['o','s','v','*','p','h','D','H']
fig=plt.figure(num=None, figsize=(11, 11), dpi=80, facecolor='w', edgecolor='k')
for i in range(X1_train.shape[0]):
    plt.plot(X1_train[i,0], X1_train[i,1],color='b',alpha=0.75 , marker=mark[y1_kmeans.labels_[i]],markersize=10.5)
plt.axis([min(X1_train[:,0])-10000,max(X1_train[:,0])+10000,min(X1_train[:,1])-10000,max(X1_train[:,1])+10000])
plt.scatter(y1_kmeans.cluster_centers_[:,0], y1_kmeans.cluster_centers_[:,1],color = 'black',linestyle='solid')
plt.Circle(xy=(y1_kmeans.cluster_centers_[:,0],y1_kmeans.cluster_centers_[:,1]), radius=max_dist_cent)
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("The "+ str(n_clus) + " Identified Clusters")
ax=fig.add_subplot(1,1,1)
for i in range(0,n_clus):
    ax.annotate("C " + str(i+1),(y1_kmeans.cluster_centers_[i,0],y1_kmeans.cluster_centers_[i,1]+0.1))
for j in range(X1_train.shape[0]):   
    ax.annotate(str(y1[j]),(X1_train[j,0], X1_train[j,1]))
plt.show()

for j in range(n_clus):
    print('Cluster: ' + str(j+1))
    for i in range(len(y1_kmeans.labels_)): 
        if y1_kmeans.labels_[i]==j:
            print(y1[i])
```
